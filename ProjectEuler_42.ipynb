## Coded Triangle Numbers
* [Link](https://projecteuler.net/problem=42)
* Using the provided file, how many of the words are 'triangle words'?
* A triangle word is a word where the sum of the values of the letters ($a=1, b=2$ etc.) is equal to a triangle number

```python
triangle_numbers = []
n = 1
triangle = 0
while n < 100:
    triangle += n
    triangle_numbers.append(triangle)
    n += 1
```

```python
letters_dict = {
    'a':1,
    'b':2,
    'c':3,
    'd':4,
    'e':5,
    'f':6,
    'g':7,
    'h':8,
    'i':9,
    'j':10,
    'k':11,
    'l':12,
    'm':13,
    'n':14,
    'o':15,
    'p':16,
    'q':17,
    'r':18,
    's':19,
    't':20,
    'u':21,
    'v':22,
    'w':23,
    'x':24,
    'y':25,
    'z':26
}
```

```python
with open("../Misc/0042_words.txt") as f:
    text = f.read()[1:-1].split('","') # 1:-1 to get rid of the opening/trailing "
```

```python
triangle_words = []
for word in text:
    word_value = 0
    for letter in word:
        word_value += letters_dict[letter.lower()]
    if word_value in triangle_numbers:
        triangle_words.append(word)
len(triangle_words)
```
